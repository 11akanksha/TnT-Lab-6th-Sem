Q1)Plot a histogram with blue color bars of size 4, and edges should be
distinguished from each other, for the dataset social_nework for the
feature estimated salary.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
plt.style.use('seaborn')
```

```python
plt.hist(df['EstimatedSalary'],
         color = 'blue',
         bins = 4,
         edgecolor = 'white'
        )
plt.title('Frequency of Estimated Salary')
plt.xlabel('Estimated salary')
plt.ylabel('Frequency')
plt.show()
```

Q4)Scatter plot age vs estimated salary on gridview
A) show regression fit line
B) Regression fit line should not be visible
C) Use * symbol to show data points without the regression fit line

```python
#A)
sns.set(style='darkgrid')
sns.regplot(x = df['Age'],y = df['EstimatedSalary'])
```

```python
# B)
sns.regplot(x = df['Age'],y = df['EstimatedSalary'],fit_reg=False)
```

```python
# C)
sns.regplot(x = df['Age'],y = df['EstimatedSalary'],fit_reg=False,marker='*')
```

Q5)Scatter plot age vs estimated salary vs purchased on dataset â€˜social
network.csvâ€™

```python
df['Purchased'].replace(1,'Yes',inplace=True)
df['Purchased'].replace(0,'No',inplace=True)
```

```python
sns.lmplot(x = 'Age',y = 'EstimatedSalary',data=df,fit_reg=False,hue='Purchased',legend=True,
          palette='Set1')
```
